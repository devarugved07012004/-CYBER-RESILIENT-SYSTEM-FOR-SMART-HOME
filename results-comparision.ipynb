```python
import matplotlib.pyplot as plt

time_sec = [66.02776336669922, 130.14866065979004, 117.68198776245117]  
accuracy_perc = [99.99, 74.42, 99.98] 

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.bar(range(len(time_sec)), time_sec, color='blue')
plt.xticks(range(len(time_sec)), ['RFE', 'PCA', 'LDA'])
plt.title('Time in Seconds')
plt.xlabel('Feature Reduction Techniques')
plt.ylabel('Time (sec)')

plt.subplot(1, 3, 2)
plt.bar(range(len(accuracy_perc)), accuracy_perc, color='green')
plt.xticks(range(len(accuracy_perc)), ['RFE', 'PCA', 'LDA'])
plt.title('Accuracy in Percentage')
plt.xlabel('Feature Reduction Techniques')
plt.ylabel('Accuracy (%)')



plt.tight_layout()
plt.show()
```
